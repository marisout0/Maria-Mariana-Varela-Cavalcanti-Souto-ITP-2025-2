```python
%%writefile main.c
#include <stdio.h>

int main() {
    int M; // Qtd de cidades (M <= 10)
    int precos[10][10];
    int X, Z; // Origem e Destino
    int i, j;

    scanf("%d", &M);

    // Leitura da matriz de preços
    for (i = 0; i < M; i++) {
        for (j = 0; j < M; j++) {
            scanf("%d", &precos[i][j]);
        }
    }

    scanf("%d %d", &X, &Z);

    // 1. Custo da rota direta (0 cidades intermediárias)
    int melhorCusto = precos[X][Z];
    int cidadeY = -1; // -1 indica que a rota direta é a melhor (ou única)

    // 2. Custo da rota com 1 cidade intermediária
    // Testa todas as cidades 'i' como possível intermediária 'Y'
    for (i = 0; i < M; i++) {
        // A intermediária não pode ser a origem nem o destino
        if (i == X || i == Z) {
            continue;
        }

        // Custo = (X -> Y) + (Y -> Z)
        int custoComY = precos[X][i] + precos[i][Z];

        // Verifica se esse novo custo é o menor encontrado
        if (custoComY < melhorCusto) {
            melhorCusto = custoComY;
            cidadeY = i; // Armazena qual foi a melhor intermediária
        }
    }

    // 3. Imprime o resultado no formato
    if (cidadeY == -1) {
        // Rota direta foi a melhor
        printf("%d-%d R$%d\n", X, Z, melhorCusto);
    } else {
        // Rota com intermediária foi a melhor
        printf("%d-%d-%d R$%d\n", X, cidadeY, Z, melhorCusto);
    }

    return 0;
}
```

```python
!gcc main.c -o main
!./main
```
