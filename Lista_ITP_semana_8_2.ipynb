```python
%%writefile chuva.c
#include <stdio.h>

int main() {
    int i, j;

    // Vetor com a quantidade de chuva (12 meses)
    int chuva[12];

    // Vetor de ponteiros para os nomes dos meses
    // Isso facilita a troca durante a ordenação
    char *meses[12] = {
        "Janeiro", "Fevereiro", "Marco", "Abril", "Maio", "Junho",
        "Julho", "Agosto", "Setembro", "Outubro", "Novembro", "Dezembro"
    };

    // Leitura dos 12 inteiros
    for (i = 0; i < 12; i++) {
        scanf("%d", &chuva[i]);
    }

    // Algoritmo de Ordenação: Selection Sort (descendente)
    // Modificado para ordem decrescente (procurando o *maior*)
    for (i = 0; i < 11; i++) {
        // Encontra o índice do maior elemento
        int indiceMaior = i;
        for (j = i + 1; j < 12; j++) {
            // Compara para achar o maior
            if (chuva[j] > chuva[indiceMaior]) {
                indiceMaior = j;
            }
        }

        // Se o maior não for o elemento 'i' atual, troca
        if (indiceMaior != i) {
            // 1. Troca a quantidade de chuva
            int auxChuva = chuva[i];
            chuva[i] = chuva[indiceMaior];
            chuva[indiceMaior] = auxChuva;

            // 2. Troca o nome do mês (ponteiro)
            char *auxMes = meses[i];
            meses[i] = meses[indiceMaior];
            meses[indiceMaior] = auxMes;
        }
    }

    // Escreve o ranking ordenado
    for (i = 0; i < 12; i++) {
        printf("%s %d\n", meses[i], chuva[i]);
    }

    return 0;
}
```
