Defini p tabuleiro como um array de caracteres e sua inicialização com espaços vazios. Também defini os símbolos para o jogador humano ('X') e para a IA ('O').

```python
%%writefile jogo_velha.c
#include <stdio.h>
#include <stdlib.h>
#include <time.h>
#include <ctype.h> // Biblioteca para funções de manipulação de caracteres

char tabuleiro[3][3]; // Vetor (array bidimensional) para representar o tabuleiro
char jogador_humano = 'X'; // Variável com tipo definido para o símbolo do jogador humano
char jogador_ia = 'O'; // Variável com tipo definido para o símbolo da IA

// Função para inicializar o tabuleiro com espaços vazios
void inicializarTabuleiro() {
    // Loop de repetição (for) para percorrer as linhas
    for (int i = 0; i < 3; i++) {
        // Loop de repetição (for) para percorrer as colunas
        for (int j = 0; j < 3; j++) {
            tabuleiro[i][j] = ' '; // Operação de atribuição
        }
    }
}

// Função para imprimir o tabuleiro
void imprimirTabuleiro() {
    // Loop de repetição (for) para percorrer as linhas
    for (int i = 0; i < 3; i++) {
        // Loop de repetição (for) para percorrer as colunas
        for (int j = 0; j < 3; j++) {
            // Comando condicional (if) para imprimir o conteúdo da célula
            printf(" %c ", tabuleiro[i][j]);
            // Comando condicional (if) para imprimir a linha divisória vertical
            if (j < 2) {
                printf("|");
            }
        }
        printf("\n"); // Quebra de linha após cada linha do tabuleiro
        // Comando condicional (if) para imprimir a linha divisória horizontal
        if (i < 2) {
            printf("---|---|---\n");
        }
    }
}

// Função para verificar se há um vencedor
char verificarVencedor() {
    // Variável com tipo definido para armazenar o símbolo do vencedor, se houver
    char vencedor = ' ';

    // Verificar linhas
    // Loop de repetição (for) para percorrer as linhas
    for (int i = 0; i < 3; i++) {
        // Comando condicional (if) para verificar se os elementos da linha são iguais e não vazios
        if (tabuleiro[i][0] == tabuleiro[i][1] && tabuleiro[i][1] == tabuleiro[i][2] && tabuleiro[i][0] != ' ') {
            vencedor = tabuleiro[i][0]; // Operação de atribuição
            return vencedor; // Retorna o vencedor encontrado
        }
    }

    // Verificar colunas
    // Loop de repetição (for) para percorrer as colunas
    for (int j = 0; j < 3; j++) {
        // Comando condicional (if) para verificar se os elementos da coluna são iguais e não vazios
        if (tabuleiro[0][j] == tabuleiro[1][j] && tabuleiro[1][j] == tabuleiro[2][j] && tabuleiro[0][j] != ' ') {
            vencedor = tabuleiro[0][j]; // Operação de atribuição
            return vencedor; // Retorna o vencedor encontrado
        }
    }

    // Verificar diagonais
    // Comando condicional (if) para verificar a diagonal principal
    if (tabuleiro[0][0] == tabuleiro[1][1] && tabuleiro[1][1] == tabuleiro[2][2] && tabuleiro[0][0] != ' ') {
        vencedor = tabuleiro[0][0]; // Operação de atribuição
        return vencedor; // Retorna o vencedor encontrado
    }

    // Comando condicional (if) para verificar a diagonal secundária
    if (tabuleiro[0][2] == tabuleiro[1][1] && tabuleiro[1][1] == tabuleiro[2][0] && tabuleiro[0][2] != ' ') {
        vencedor = tabuleiro[0][2]; // Operação de atribuição
        return vencedor; // Retorna o vencedor encontrado
    }

    return vencedor; // Retorna ' ' se não houver vencedor
}

// Função para verificar se há empate
int verificarEmpate() {
    // Loop de repetição (for) para percorrer as linhas
    for (int i = 0; i < 3; i++) {
        // Loop de repetição (for) para percorrer as colunas
        for (int j = 0; j < 3; j++) {
            // Comando condicional (if) para verificar se alguma célula está vazia
            if (tabuleiro[i][j] == ' ') {
                return 0; // Retorna 0 se encontrar uma célula vazia (não é empate)
            }
        }
    }
    return 1; // Retorna 1 se o tabuleiro estiver cheio (é empate)
}

// Função para a jogada do jogador humano
void jogadaHumano() {
    int linha, coluna; // Variáveis com tipo definido para armazenar a linha e coluna da jogada

    // Loop de repetição (do-while) para garantir que a jogada seja válida
    do {
        printf("Sua vez (%c). Digite a linha (0-2) e a coluna (0-2): ", jogador_humano); // Imprime mensagem para o jogador
        scanf("%d %d", &linha, &coluna); // Lê a entrada do usuário

        // Comando condicional (if) para verificar se a jogada está dentro dos limites do tabuleiro
        if (linha < 0 || linha > 2 || coluna < 0 || coluna > 2) {
            printf("Jogada inválida. Fora dos limites do tabuleiro.\n"); // Imprime mensagem de erro
        }
        // Comando condicional (else if) para verificar se a célula já está ocupada
        else if (tabuleiro[linha][coluna] != ' ') {
            printf("Jogada inválida. Posição já ocupada.\n"); // Imprime mensagem de erro
        }
        // Comando condicional (else) para realizar a jogada se for válida
        else {
            tabuleiro[linha][coluna] = jogador_humano; // Operação de atribuição
            break; // Sai do loop se a jogada for válida
        }
    } while (1); // Continua o loop até que uma jogada válida seja feita
}

// Função para a jogada da IA básica
void jogadaIA() {
    int linha, coluna; // Variáveis com tipo definido para armazenar a linha e coluna da jogada da IA

    // Inicializa a semente para números aleatórios
    srand(time(0)); // Chamada de função para inicializar a semente do gerador de números aleatórios

    // Loop de repetição (do-while) para encontrar uma posição vazia aleatória
    do {
        linha = rand() % 3; // Operação aritmética para gerar número aleatório entre 0 e 2 para a linha
        coluna = rand() % 3; // Operação aritmética para gerar número aleatório entre 0 e 2 para a coluna
    } while (tabuleiro[linha][coluna] != ' '); // Continua o loop enquanto a posição estiver ocupada

    tabuleiro[linha][coluna] = jogador_ia; // Operação de atribuição: a IA faz sua jogada
    printf("Vez da IA (%c):\n", jogador_ia); // Imprime mensagem indicando a jogada da IA
}


int main() {
    inicializarTabuleiro();
    char vencedor = ' '; // Variável para armazenar o vencedor
    int empate = 0; // Variável para verificar empate

    // Loop principal do jogo
    // Loop de repetição (while) que continua enquanto não houver vencedor e não houver empate
    while (vencedor == ' ' && empate == 0) {
        imprimirTabuleiro(); // Imprime o tabuleiro antes da jogada do humano
        jogadaHumano(); // Chama a função para a jogada do humano

        vencedor = verificarVencedor(); // Verifica se o humano venceu
        // Comando condicional (if) para verificar se há um vencedor ou empate após a jogada do humano
        if (vencedor != ' ' || verificarEmpate()) {
            break; // Sai do loop se houver vencedor ou empate
        }

        jogadaIA(); // Chama a função para a jogada da IA
        vencedor = verificarVencedor(); // Verifica se a IA venceu
        empate = verificarEmpate(); // Verifica se houve empate
    }

    imprimirTabuleiro(); // Imprime o tabuleiro final

    // Comando condicional (if) para verificar o resultado do jogo
    if (vencedor == jogador_humano) {
        printf("Parabéns! Você venceu!\n"); // Imprime mensagem de vitória do humano
    } else if (vencedor == jogador_ia) {
        printf("A IA venceu. Tente novamente!\n"); // Imprime mensagem de vitória da IA
    } else {
        printf("O jogo terminou em empate!\n"); // Imprime mensagem de empate
    }

    return 0;
}
```
