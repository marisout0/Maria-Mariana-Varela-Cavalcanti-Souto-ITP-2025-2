```python
%%writefile campo_agricola.c
#include <stdio.h>

int main() {
    int M, N;
    // Limites M <= 10, N <= 10
    int campo[10][10];

    // Matriz auxiliar para marcar setores irrigados (evitar contagem dupla)
    // Inicializa toda a matriz com 0 (falso)
    int irrigado[10][10] = {0};

    int i, j;

    // Leitura das dimensões M e N
    scanf("%d %d", &M, &N);

    // Leitura da matriz 'campo'
    // Usa repetições aninhadas para percorrer a matriz
    for (i = 0; i < M; i++) {
        for (j = 0; j < N; j++) {
            scanf("%d", &campo[i][j]);
        }
    }

    // 1. Encontra os irrigadores (2) e marca os setores que eles irrigam
    for (i = 0; i < M; i++) {
        for (j = 0; j < N; j++) {
            if (campo[i][j] == 2) {
                // O problema garante que irrigadores não estão na borda
                // Portanto, não precisamos verificar os limites (i-1, i+1, j-1, j+1)

                // Marca os 4 vizinhos como irrigados
                irrigado[i - 1][j] = 1; // Norte
                irrigado[i + 1][j] = 1; // Sul
                irrigado[i][j - 1] = 1; // Oeste
                irrigado[i][j + 1] = 1; // Leste
            }
        }
    }

    int ferteis_irrigados = 0;
    int ferteis_nao_irrigados = 0;

    // 2. Contabiliza os setores férteis (1)
    for (i = 0; i < M; i++) {
        for (j = 0; j < N; j++) {
            // Verifica se o setor é fértil (valor 1)
            if (campo[i][j] == 1) {
                if (irrigado[i][j] == 1) {
                    ferteis_irrigados++;
                } else {
                    ferteis_nao_irrigados++;
                }
            }
        }
    }

    // Escreve os resultados
    printf("%d %d\n", ferteis_irrigados, ferteis_nao_irrigados);

    return 0;
}
```
