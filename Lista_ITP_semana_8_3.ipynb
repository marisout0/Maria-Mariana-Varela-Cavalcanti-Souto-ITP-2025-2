```python
%%writefile "esse sim é o piloto.c"

#include <stdio.h>

int main() {
    int n; // Qtd de pilotos (1 <= n <= 20)
    scanf("%d", &n);

    // Arrays para guardar as posições (0 a n-1) de cada piloto (1 a n)
    // Usamos n+1 para indexar diretamente pelo ID do piloto (1 a n)
    int posLargada[21];
    int posChegada[21];

    int pilotoID;
    int i;

    // 1. Lê a ordem de largada e armazena a posição (i)
    //    no índice correspondente ao ID do piloto.
    for (i = 0; i < n; i++) {
        scanf("%d", &pilotoID); // Lê o ID do piloto na posição i
        posLargada[pilotoID] = i; // Posição 'i' (0 = 1º, 1 = 2º, ...)
    }

    // 2. Lê a ordem de chegada e armazena a posição (i)
    for (i = 0; i < n; i++) {
        scanf("%d", &pilotoID);
        posChegada[pilotoID] = i;
    }

    int maxGanho = 0;
    int pilotoVencedor = 0; // Guarda o ID do piloto
    int houveEmpate = 0;    // Flag para caso de empate

    // 3. Calcula o ganho para cada piloto
    for (pilotoID = 1; pilotoID <= n; pilotoID++) {
        int ganho = posLargada[pilotoID] - posChegada[pilotoID];

        if (ganho > maxGanho) {
            maxGanho = ganho;
            pilotoVencedor = pilotoID;
            houveEmpate = 0; // Novo máximo, reseta o empate
        } else if (ganho == maxGanho && maxGanho > 0) {
            // Empatou com o ganho máximo atual
            houveEmpate = 1;
        }
    }

    // 4. Imprime o resultado
    // Se ninguém ganhou posições (maxGanho == 0) ou se houve empate
    if (maxGanho == 0 || houveEmpate == 1) {
        printf("empate\n");
    } else {
        printf("%d\n", pilotoVencedor); // Imprime o ID
    }

    return 0;
}
```
