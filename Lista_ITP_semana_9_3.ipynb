```python
%%writefile /content/empates.c
#include <stdio.h>

int main() {
    int M; // Qtd de times (M <= 20)
    int placares[20][20];
    int i, j;
    int total_empates = 0;

    scanf("%d", &M);

    // Leitura da matriz M x M
    for (i = 0; i < M; i++) {
        for (j = 0; j < M; j++) {
            scanf("%d", &placares[i][j]);
        }
    }

    // Conta os empates
    // Cada time enfrenta os outros uma única vez
    // Para evitar contar duas vezes (i vs j) e (j vs i),
    // percorremos apenas o triângulo superior da matriz (onde j > i).
    for (i = 0; i < M; i++) {
        for (j = i + 1; j < M; j++) {
            // Verifica o placar i vs j
            if (placares[i][j] == placares[j][i]) {
                total_empates++;
            }
        }
    }

    // Escreve o número de partidas empatadas
    printf("%d\n", total_empates);

    return 0;
}
```
