```python
%%writefile sopa.c
#include <stdio.h>
#include <stdlib.h> // Para malloc() e free()
#include <string.h> // Para strlen()

/**
 * Mistura duas strings, alternando seus caracteres.
 * Retorna uma nova string alocada dinamicamente.
 *
 */
char* misturar(char *str1, char *str2) {
    int len1 = strlen(str1);
    int len2 = strlen(str2);
    int len_nova = len1 + len2;

    // 1. Aloca memória para a nova string (tamanho total + 1 para '\0')
    char *nova_str = (char*) malloc((len_nova + 1) * sizeof(char));

    int i = 0; // índice para str1
    int j = 0; // índice para str2
    int k = 0; // índice para nova_str

    // 2. Alterna os caracteres enquanto ambas as strings tiverem letras
    while (i < len1 && j < len2) {
        nova_str[k++] = str1[i++];
        nova_str[k++] = str2[j++];
    }

    // 3. Adiciona o restante da str1 (se houver)
    while (i < len1) {
        nova_str[k++] = str1[i++];
    }

    // 4. Adiciona o restante da str2 (se houver)
    while (j < len2) {
        nova_str[k++] = str2[j++];
    }

    // 5. Adiciona o caractere nulo no final
    nova_str[k] = '\0';

    return nova_str; // Retorna o ponteiro para a nova string
}

int main() {
    // Strings de até 100 caracteres +1 para '\0'
    char str1[101];
    char str2[101];

    // Lê as duas strings (uma por linha, com ou sem espaços)
    // Usamos 'fgets' para maior segurança
    fgets(str1, sizeof(str1), stdin);
    fgets(str2, sizeof(str2), stdin);

    // Remove o newline adicionado por fgets
    str1[strcspn(str1, "\n")] = 0;
    str2[strcspn(str2, "\n")] = 0;


    // Chama a função para misturar
    char *resultado = misturar(str1, str2);

    // Imprime o resultado
    printf("%s\n", resultado);

    // Libera a memória alocada dinamicamente pela função 'misturar'
    free(resultado);

    return 0;
}
```
