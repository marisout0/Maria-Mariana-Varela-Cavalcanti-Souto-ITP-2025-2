```python
%%writefile estou_com_sorte.c

#include <stdio.h>

int main() {
    int m, n;

    // Lê M (qtd sorteados) e N (qtd aposta)
    scanf("%d %d", &m, &n);

    // Limites máximos dados: M <= 30, N <= 50
    int sorteados[50];
    int aposta[50];
    int i, j;
    int acertos = 0;

    // Lê os M números sorteados
    for (i = 0; i < m; i++) {
        scanf("%d", &sorteados[i]);
    }

    // Lê os N números da aposta
    for (i = 0; i < n; i++) {
        scanf("%d", &aposta[i]);
    }

    // Utiliza repetições aninhadas para comparar cada número da
    // aposta com cada número sorteado.
    for (i = 0; i < n; i++) { // Loop externo: percorre aposta
        for (j = 0; j < m; j++) { // Loop interno: percorre os sorteados
            if (aposta[i] == sorteados[j]) {
                acertos++;
                // O número não aparece mais de uma vez no sorteio
                // Então podemos parar de procurar (break)
                break;
            }
        }
    }

    // Escreve o total de acertos
    printf("%d\n", acertos);

    return 0;
}
```
