```python
%%writefile amigos.c

#include <stdio.h>

int main() {
    int M; // Qtd de usuários (M <= 100)
    int amigos[100][100];
    int x; // Usuário para quem vamos sugerir
    int i, j, k;

    scanf("%d", &M);

    // Leitura da matriz M x M
    for (i = 0; i < M; i++) {
        for (j = 0; j < M; j++) {
            scanf("%1d", &amigos[i][j]); // Leitura de dígitos individuais
        }
    }

    scanf("%d", &x); // Leitura do usuário 'x'

    // Loop para testar cada usuário 'j' como possível sugestão para 'x'
    for (j = 0; j < M; j++) {
        // Não pode sugerir a si mesmo (amigos[ii] = 0 de qualquer forma)
        if (x == j) {
            continue;
        }

        // Condição 1: 'x' e 'j' NÃO são amigos
        if (amigos[x][j] == 0) {

            int amigo_em_comum = 0; // Flag para verificar amigo em comum

            // Condição 2: Procurar um amigo em comum 'k'
            for (k = 0; k < M; k++) {
                // 'x' é amigo de 'k' E 'j' é amigo de 'k'
                if (amigos[x][k] == 1 && amigos[j][k] == 1) {
                    amigo_em_comum = 1;
                    // Encontrou UM amigo em comum, já satisfaz a condição
                    break;
                }
            }

            // Se ambas as condições são verdadeiras, imprime a sugestão
            if (amigo_em_comum) {
                // Como o loop 'j' vai de 0 a M-1, a saída
                // já será em ordem numérica crescente
                printf("%d ", j);
            }
        }
    }
    printf("\n"); // Quebra de linha no final

    return 0;
}
```
