Exercío 5

 Escreva um programa que leia um número n e escreva o padrão na tela.

```python
%%writefile pattern.c
#include <stdio.h>

int main() {
    int n;
    int contador = 1; // Variável para controlar os números a serem impressos.

    printf("Digite o numero de linhas (n): ");
    scanf("%d", &n);

    // Laço externo para as linhas.
    for (int i = 1; i <= n; i++) {
        // Laço interno para os números em cada linha. O número de colunas é igual ao número da linha.
        for (int j = 1; j <= i; j++) {
            printf("%d ", contador);
            contador++; // Incrementa o contador a cada número impresso.
        }
        printf("\n"); // Quebra de linha para a próxima linha do padrão.
    }

    return 0;
}
```

Escreva um programa que escreva na tela todos os trios pitagóricos onde o maior número é menor que 1000.

```python
%%writefile pythagorean.c
#include <stdio.h>

int main() {
    int a, b, c;
    printf("Trios pitagoricos com o maior numero menor que 1000:\n");

    // Usa 3 laços para testar todas as combinações de a, b e c até 1000.
    // Otimiza o código para evitar combinações repetidas, garantindo a < b < c.
    for (a = 1; a < 1000; a++) {
        for (b = a; b < 1000; b++) {
            for (c = b; c < 1000; c++) {
                // Verifica a condição do trio pitagórico.
                if (a * a + b * b == c * c) {
                    printf("(%d, %d, %d)\n", a, b, c);
                }
            }
        }
    }

    return 0;
}
```

Escreva um programa que leia dois números inteiros a e b...e escreva todos os números primos entre a e b.

```python
%%writefile prime_numbers.c
#include <stdio.h>
#include <stdbool.h> // Para usar o tipo de dado bool

int main() {
    int a, b;
    bool eh_primo;

    // Garante que a > b.
    do {
        printf("Digite dois numeros inteiros (a e b, onde a > b): ");
        scanf("%d %d", &a, &b);
    } while (a <= b);

    printf("Numeros primos entre %d e %d:\n", b, a);

    // Troca os valores se necessário para garantir que o laço vá do menor para o maior.
    int menor = b, maior = a;

    // Laço para verificar cada número no intervalo.
    for (int num = menor; num <= maior; num++) {
        // Ignora 0 e 1, que não são primos.
        if (num <= 1) {
            continue;
        }

        eh_primo = true; // Assume que o número é primo.
        // Laço para testar a primalidade.
        for (int i = 2; i * i <= num; i++) {
            if (num % i == 0) {
                eh_primo = false;
                break; // Se um divisor for encontrado, não é primo.
            }
        }

        // Se a flag 'eh_primo' ainda for verdadeira, o número é primo.
        if (eh_primo) {
            printf("%d\n", num);
        }
    }

    return 0;
}
```

Escreva um programa que escreva na tela quanto resta em cada uma das possibilidades

```python
%%writefile calculate_remaining.c
#include <stdio.h>

int main() {
    float preco1, preco2, quantia_disponivel;
    int i, j;
    float total_gasto, restante;

    printf("Digite o preco do primeiro fornecedor: ");
    scanf("%f", &preco1);

    printf("Digite o preco do segundo fornecedor: ");
    scanf("%f", &preco2);

    printf("Digite a quantia disponivel: ");
    scanf("%f", &quantia_disponivel);

    // O laço externo itera sobre a quantidade do fornecedor 1 (0 a 10).
    for (i = 0; i <= 10; i++) {
        // O laço interno itera sobre a quantidade do fornecedor 2 (0 a 10).
        for (j = 0; j <= 10; j++) {
            total_gasto = (i * preco1) + (j * preco2);

            // Verifica se a compra é possível.
            if (total_gasto <= quantia_disponivel) {
                restante = quantia_disponivel - total_gasto;
                printf("Comprando %d do primeiro e %d do segundo resta: %.2f\n", i, j, restante);
            }
        }
    }

    return 0;
}
```

Modifique o programa da questão anterior para que escreva na tela quanto comprar de cada fornecedor para que reste a menor quantia

```python
%%writefile calculate_remaining.c
#include <stdio.h>

int main() {
    float preco1, preco2, quantia_disponivel;
    int i, j, melhor_i = 0, melhor_j = 0;
    float total_gasto, menor_restante;

    printf("Digite o preco do primeiro fornecedor: ");
    scanf("%f", &preco1);

    printf("Digite o preco do segundo fornecedor: ");
    scanf("%f", &preco2);

    printf("Digite a quantia disponivel: ");
    scanf("%f", &quantia_disponivel);

    // Inicializa o menor_restante com um valor impossível, para que a primeira compra válida seja registrada.
    menor_restante = quantia_disponivel + 1;

    // Laços aninhados para testar todas as combinações.
    for (i = 0; i <= 10; i++) {
        for (j = 0; j <= 10; j++) {
            total_gasto = (i * preco1) + (j * preco2);

            if (total_gasto <= quantia_disponivel) {
                float restante = quantia_disponivel - total_gasto;

                // Se o restante atual for menor que o menor restante registrado, atualiza.
                if (restante < menor_restante) {
                    menor_restante = restante;
                    melhor_i = i;
                    melhor_j = j;
                }
            }
        }
    }

    printf("Resta menos comprando %d do primeiro e %d do segundo\n", melhor_i, melhor_j);

    return 0;
}
```

A melhor promoção

```python
%%writefile best_promotion.c
#include <stdio.h>

int main() {
    float p, q;
    float lucro, melhor_razao = 0.0;
    int melhor_x = 0, melhor_y = 0;

    printf("Digite o preco de venda p: ");
    scanf("%f", &p);

    printf("Digite o preco de producao q: ");
    scanf("%f", &q);

    // x representa as unidades que o cliente leva, e y as que ele paga.
    // O pacote tem no máximo 10 itens, ou seja, x <= 10.
    for (int x = 2; x <= 10; x++) {
        // O cliente paga y unidades, onde y é menor que x.
        for (int y = 1; y < x; y++) {
            // Calcula o lucro: (preço de venda * unidades pagas) - (preço de produção * unidades levadas).
            lucro = (y * p) - (x * q);

            // Se não houver prejuízo (lucro >= 0).
            if (lucro >= 0) {
                // Calcula a razão x/y.
                float razao = (float)x / y;

                // Se a razão atual for a melhor encontrada até agora.
                if (razao > melhor_razao) {
                    melhor_razao = razao;
                    melhor_x = x;
                    melhor_y = y;
                }
            }
        }
    }

    // Se uma promoção válida foi encontrada.
    if (melhor_x > 0) {
        printf("A melhor promocao eh: leve %d pague %d\n", melhor_x, melhor_y);
    } else {
        printf("Nao ha promocao viavel sem prejuizo.\n");
    }

    return 0;
}
```
