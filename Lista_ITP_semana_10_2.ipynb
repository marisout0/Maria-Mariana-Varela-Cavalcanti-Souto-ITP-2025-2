```python
%%writefile ocorrencia_de_vetor.c

#include <stdio.h>
#include <stdlib.h> // Para malloc(), free(), NULL

/**
 * Busca todas as ocorrências de 'valor' no vetor 'v' de tamanho 'n'.
 * Retorna um vetor dinâmico com os índices.
 * Armazena a quantidade de ocorrências no ponteiro 'qtd'.
 * Retorna NULL se não houver ocorrências.
 */
int* buscaNoVetor(int *v, int n, int valor, int *qtd) {
    int i;
    int contagem = 0;

    // 1. Primeira passagem: contar quantas ocorrências existem
    for (i = 0; i < n; i++) {
        if (v[i] == valor) {
            contagem++;
        }
    }

    // 2. Armazena a contagem no ponteiro de retorno indireto
    *qtd = contagem;

    // 3. Se não houver ocorrências, retorna NULL
    if (contagem == 0) {
        return NULL;
    }

    // 4. Aloca memória para o vetor de índices
    int *indices = (int*) malloc(contagem * sizeof(int));
    if (indices == NULL) { // Boa prática: verificar falha no malloc
        *qtd = 0;
        return NULL;
    }

    // 5. Segunda passagem: preencher o vetor de índices
    int idx_indices = 0; // Índice para o vetor 'indices'
    for (i = 0; i < n; i++) {
        if (v[i] == valor) {
            indices[idx_indices] = i;
            idx_indices++;
        }
    }

    return indices;
}


int main() {
    int n, valor, i;

    // 1. Ler N (tamanho do vetor)
    scanf("%d", &n);

    // 2. Alocar vetor 'v' dinamicamente
    int *v = (int*) malloc(n * sizeof(int));

    // 3. Ler N inteiros para 'v'
    for (i = 0; i < n; i++) {
        scanf("%d", &v[i]);
    }

    // 4. Ler o 'valor' a ser buscado
    scanf("%d", &valor);

    int qtdOcorrencias = 0; // Variável para guardar a quantidade

    // 5. Chamar a função
    // Passamos o ENDEREÇO de qtdOcorrencias (&)
    int *indices = buscaNoVetor(v, n, valor, &qtdOcorrencias);

    // 6. Imprimir o resultado
    if (indices == NULL) {
        printf("Nenhuma ocorrencia\n");
    } else {
        // Qtd de ocorrências
        printf("%d\n", qtdOcorrencias);

        // Índices
        for (i = 0; i < qtdOcorrencias; i++) {
            printf("%d", indices[i]);
            if (i < qtdOcorrencias - 1) {
                printf(" ");
            }
        }
        printf("\n");
    }

    // 7. Liberar memória alocada
    free(v);
    if (indices != NULL) {
        free(indices);
    }

    return 0;
}
```
