```python
%%writefile openmeet.c

#include <stdio.h>
#include <string.h> // Para strtok(), strcmp(), strlen()

// Função para verificar se uma string (token) é uma data no formato DD/MM
int ehData(char *token) {
    // Formato DD/MM tem 5 caracteres e '/' na posição 2
    if (strlen(token) == 5 && token[2] == '/') {
        return 1;
    }
    return 0;
}

int main() {
    // Nome do aluno (máx 10 chars) +1 para '\0'
    char nomeAluno[11];

    // String de frequências (máx 1000 chars) +1 para '\0'
    // Precisamos de buffer extra para o nome do aluno que vem na mesma linha.
    // Vamos ler a linha inteira primeiro.
    char linhaInteira;

    char *token;
    int totalAulas = 0;
    int presencas = 0;

    // Lê o nome do aluno
    scanf("%s", nomeAluno);

    // Lê o RESTO da linha (a string de frequências)
    // Usamos 'gets' conforme sugerido no slide 13
    // (Consome o espaço deixado pelo scanf antes)
    char espaco;
    scanf("%c", &espaco); // Consome o espaço entre o nome e o resto
    fgets(linhaInteira, sizeof(linhaInteira), stdin);


    // Usamos strtok() para "quebrar" a string pelos espaços
    // A biblioteca string.h é permitida
    token = strtok(linhaInteira, " ");

    while (token != NULL) {
        // Verifica se o token é uma data
        if (ehData(token)) {
            totalAulas++;
        }
        // Verifica se o token é o nome do aluno
        // strcmp() compara duas strings
        else if (strcmp(token, nomeAluno) == 0) {
            presencas++;
        }

        // Pega o próximo token
        token = strtok(NULL, " ");
    }

    // Calcula e imprime as ausências
    printf("%d\n", totalAulas - presencas);

    return 0;
}
```
